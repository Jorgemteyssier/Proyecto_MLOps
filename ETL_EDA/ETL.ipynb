```python
#Importamos las librerías y/o frameworks a usar

import pandas as pd
```

```python
#Definimos los data frames abriendo los Datasets que nos proporcionaron 
 
df_movies=pd.read_csv('Data\movies_dataset.csv')
df_credits=pd.read_csv('Data\credits.csv')

#Ajustamos las opciones del df para poder ver todas las columnas 

pd.set_option('display.max_columns', None)
```

```python
#Iniciamos con el punto 6 para aligerar la data eliminando datos que no vamos a usar 
#6. Eliminar las columnas que no serán utilizadas, video,imdb_id,adult,original_title,poster_path y homepage.

df_movies.drop(columns=['video','imdb_id','adult','original_title','poster_path','homepage'],inplace=True)

#Usamos inplace=True ya que no usaremos esas columnas por lo que se pueden eliminar del df original
```

```python
#Vamos a seguir trabajndo con el df_movies ya que es el que mas transformaciones requiere
#obsevamos los datos y la cantidad y tipo de estos, no se centraremos en la columna id que es la que usaremos como clave durante el etl
df_movies.info()
```

```python
#Vamos a eliminar todos los valores nulos, duplicados y pasaremos todo a valores de tipo int 

#Elimina los valores nulos
df_movies=df_movies.dropna(subset=['id'])
```

```python
#Checa y elimina los valores duplicados dejando solo uno de cada uno
valores=df_movies['id'].value_counts()
valores_a_eliminar=valores[valores>1].index
df_movies=df_movies[~df_movies['id'].isin(valores_a_eliminar)]
```

```python
#pasamos los datos a valores numericos y volvemos a eliminar nulos por si un str se volvio nulo al pasar a numerico y despues los convertimos a enteros usando el entero con menor tamaño posible 
#con el fin de reducir el archivo y la memoria lo más que se pueda
df_movies['id']=pd.to_numeric(df_movies['id'], errors='coerce',downcast='integer')
df_movies=df_movies.dropna(subset=['id'])
df_movies['id']=df_movies['id'].astype('int')
```

```python
#Vamos a hacer lo mismo con df_credits
df_credits=df_credits.dropna(subset=['id'])

valores_credits=df_credits['id'].value_counts()
valores_a_eliminar_credits=valores_credits[valores_credits>1].index
df_credits=df_credits[~df_credits['id'].isin(valores_a_eliminar_credits)]

df_credits['id']=pd.to_numeric(df_credits['id'], errors='coerce',downcast='integer')
df_credits=df_credits.dropna(subset=['id'])
df_credits['id']=df_credits['id'].astype('int')


#Con esto tenemos los dos df con ambas columnas id limpias las cuales seran nuestras claves y con las columnas que no vamos a utilizar eliminadas
```

```python
#regresamos al punto 1 despues de eliminar columnas y limpiar la columna que usaremos como clave y procedmos a desanidar 
#1. Algunos campos, como belongs_to_collection, production_companies y otros (ver diccionario de datos) están anidados, esto es o bien tienen un diccionario o una lista como valores en cada fila,
# ¡deberán desanidarlos para poder y unirlos al dataset de nuevo hacer alguna de las consultas de la API! O bien buscar la manera de acceder a esos datos sin desanidarlos.

df_movies.head()
```

```python
#Vemos que 'belongs_to_collection', 'genres', 'production_companies', 'production_countries', 'spoken_languages' tienen datos anidados
#Importamos ast para tratar los str
import ast

#Inicismos con 'belongs_to_collection'
df_movies['belongs_to_collection'].head()
```

```python
#Hacemos un nuevo df para la columna, agregamos la columna id ya que sera la clave para hacer merge o poder consultar
#tenemos que hacer .dropna ya que a id le hicimos las transformaciones pertinente pero belongs to collection no, por lo que si hay datos nulos no va a coincidir el numero de filas
df_bc = df_movies[['id', 'belongs_to_collection']].dropna()
```

```python
#Con ayuda de ast vamos a pasar los str de belong collection a dict
df_bc['belongs_to_collection'] = df_bc['belongs_to_collection'].apply(ast.literal_eval)
```

```python
df_bc.head()
```

```python
# Usamos pd.json_normalize para desanidar la columna de belong coleection y crear un df con las columnas desanidadas
df_bc_desanidado = pd.json_normalize(df_bc['belongs_to_collection'])
```

```python
df_bc_desanidado.head()
```

```python
#observamos que de este df solo vamos a ocupar el nombre para las funciones en la api y el modelo por lo que las demas columnas las vamos a eliminar para aligerar la data
df_bc_desanidado.drop(columns=['id','poster_path','backdrop_path'],inplace=True)
```

```python
#Unimos la columna desanidada con la de id concatenandolas
df_bc_desanidado=pd.concat([df_bc['id'],df_bc_desanidado],axis=1)
```

```python
df_bc_desanidado.head()
```

```python
#Pasamos a genres 
df_movies['genres'].head()
```

```python
#Realizamos lois mismos pasos

df_genres = df_movies[['id', 'genres']].dropna()
df_genres['genres'] = df_genres['genres'].apply(ast.literal_eval)
df_genres_desanidado = pd.json_normalize(df_genres['genres'])
```

```python
#Observamos que al desanidar quedan diciconarios ya que la columna tenia listas con diccionarios adentro podemos ver que llega a haber hasta 7 elementos en la lista y que varios estan vacios
#Para este proyecto nos interesa saber el genero de cada pelicula por lo que los elemntos vacio y el id de cada genero no nos importan
df_genres_desanidado.head()
```

```python
#Creamos una funcion que nos permita acceder a cada elemnto y agruparlos como una sola lista de generos
def extraer_genero(x):
    return[genre['name'] for genre in x ]
```

```python
#La aplicamos a toda la columna de genero de df_genres 
df_genres['genres']=df_genres['genres'].apply(extraer_genero)
```

```python
df_genres.head()
```

```python
#Para mayor comprension lo guardamos en el df_genres_desanidado
df_genres_desanidado=df_genres
```

```python
#Realizamos lois mismos pasos

df_pc = df_movies[['id', 'production_companies']].dropna()
df_pc['production_companies'] = df_pc['production_companies'].apply(ast.literal_eval)
df_pc_desanidado = pd.json_normalize(df_pc['production_companies'])
```

```python
#Observamos que es el mismo problema que con el genero, tenemos una lista con muchos elementos y son diccionarios de los cuales solo nos importa el nombre
#Por lo que aplicaremos la misma funcion de extraccion pero con production_companmies
#De haber sabido que se presentaria el mismo problema lo más optimo era crear una funcion generica para cualquier columna 
df_pc_desanidado.head()
```

```python
#Se observa que es similar a las columnas anteriores por lo que haremos el mismo desanidado 
df_sl = df_movies[['id', 'spoken_languages']].dropna()
df_sl['spoken_languages'] = df_sl['spoken_languages'].apply(ast.literal_eval)
df_sl_desanidado = pd.json_normalize(df_sl['spoken_languages'])
```

```python
#Ya tengo la data que nos interesa de cada columna anidada y relacionada con la clave id
#Voy a hacer un merge de todos los df desanidados con el df_movies 

#Creamos una lista con todos los df
lista_df=[df_movies,df_bc_desanidado,df_genres_desanidado,df_pc_desanidado,df_pcountries_desanidado,df_sl_desanidado]

#Creamos el df final y lo inicializamos
df_movies_limpio=lista_df[0]

#Iteramos para que se haga merge en todos 
for a in lista_df[1:]:
    df_movies_limpio=pd.merge(df_movies_limpio,a, on='id', how='left')
```

```python
df_movies_limpio.head()
```

```python
#Eliminamos las columnas anidadas que ya no ocupamos
df_movies_limpio.drop(columns=['belongs_to_collection', 'genres_x', 'production_companies_x', 'production_countries_x', 'spoken_languages_x'],inplace=True)
```

```python
#Voy a renombrar las columnas a las que se les agrego el sufijo_y ya que estaban duplicadas y la de name a collection para mayor claridad
df_movies_limpio.rename(columns={'genres_y':'genres', 'production_companies_y':'production_companies', 'production_countries_y':'production_countries', 'spoken_languages_y':'spoken_languages','name':'collection'}, inplace=True)
```

```python
#Sigo al punto 2
#2. Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0.
df_movies_limpio.info()
```

```python
#Observo que budget no esta en un dato numerico
df_movies_limpio['budget']=pd.to_numeric(df_movies_limpio['budget'], errors='coerce',downcast='integer')

#Cambio los nulos por 0
df_movies_limpio['budget'].fillna(0,inplace=True)
df_movies_limpio['revenue'].fillna(0,inplace=True)
```

```python
#3. Los valores nulos del campo release date deben eliminarse.
df_movies_limpio=df_movies_limpio.dropna(subset=['release_date'])
```

```python
#4. De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

#Convierto a datetime
df_movies_limpio['release_date']=pd.to_datetime(df_movies_limpio['release_date'], errors='coerce')
```

```python
df_movies_limpio['release_date'].dtype
```

```python
#Extraemos el año y lo ponemos en la columna release_year
df_movies_limpio['release_year']=df_movies_limpio['release_date'].dt.year
```

```python
#Le doy el formato requerido esto lo hago al final ya que me di cuenta que si lo hacia antes el tipo de dato se cambiaba a object y no podia extraer el año con .dt.year
df_movies_limpio['release_date']=df_movies_limpio['release_date'].dt.strftime('%Y-%m-%d')
```

```python
df_movies_limpio.head()
```

```python
#5. Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget
#cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

df_movies_limpio['return']=df_movies_limpio.apply(lambda q: q['revenue'] / q['budget'] if q['budget'] != 0 else 0,axis=1)
```

```python
#Pasamos los object a list

df_credits['cast'] = df_credits['cast'].apply(ast.literal_eval)
df_credits['crew'] = df_credits['crew'].apply(ast.literal_eval)
```

```python
#Desanido los datos como en las columnas de movies usando una funcion para extraerlos
def extraer_cast(x):
    return[cast['name'] for cast in x ]

#Leyendo las funciones de la api veo que solo ocupamos los nombres de los actores por lo que para no realizar tantos df  agregamos directo la nueva columna al df original así evitamos hacer los merge 

df_credits['cast_members']=df_credits['cast'].apply(extraer_cast)
```

```python
#Hago lo mismo con crew solo me voy a quedar con el director ya que es lo unico que ocupo para la api
def director(x):
    for director in x:
        if director['job'] == 'Director':
            return director['name']
```

```python
df_credits['director']=df_credits['crew'].apply(director)
```
